```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True).astype(int)
dataset
```

```python
dataset.drop("User ID",axis = 1)
dataset
```

```python
indep=dataset[["Age","EstimatedSalary","Gender_Male"]]
dep=dataset["Purchased"]
dataset["Purchased"].value_counts()
```

```python
#Splitting train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(indep,dep,test_size=0.30,random_state=0)
print(x_train)
print(x_test)
```

```python
#Model Creation using GaussianNB
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(x_train,y_train)
#Evaluation Mertics
y_predict = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
#To get report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_predict)
print(cm)
print(cr)
```

```python
#Model Creation using MultinomialNB
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
classifier.fit(x_train,y_train)
#Evaluation Mertics
y_predict = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
#To get report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_predict)
print(cm)
print(cr)
```

```python
#Model Creation using ComplementNB
from sklearn.naive_bayes import ComplementNB
classifier = ComplementNB()
classifier.fit(x_train,y_train)
#Evaluation Mertics
y_predict = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
#To get report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_predict)
print(cm)
print(cr)
```

```python
#Model Creation using BernoulliNB
from sklearn.naive_bayes import BernoulliNB
classifier = BernoulliNB()
classifier.fit(x_train,y_train)
#Evaluation Mertics
y_predict = classifier.predict(x_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
#To get report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_predict)
print(cm)
print(cr)
```

```python
#Model Creation using CategoricalNB
from sklearn.naive_bayes import CategoricalNB
classifier = CategoricalNB()
classifier.fit(x_train,y_train)
y_predict = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
#To get report
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_predict)
print(cm)
print(cr)
```
